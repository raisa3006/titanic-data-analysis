```python
#Extracting salutatin from Name(Format: Last Name, First Title, First Name)
def title(name):
    temp_1 = name.split(',') # Split by (,)
    temp_2 = temp_1[1].split('.')[0] # Split by (.)
    temp_3 = temp_2.strip() # Remove white space
    return temp_3
    
```

```python
train_df["Title"] = train_df["Name"].apply(title)
```

```python
def new_title(title):
    if title == 'Mr' or title == 'Capt' or title == 'Don' or title == 'Dr' or title == 'Jonkheer' or title     == 'Major' or  title == 'Rev' or title == 'Sir' or title == 'Col':
        return 'Mr'
    elif title == "Mrs" or title == "Lady" or title == "Mme" or title == "Ms" or title == "the Countess":
        return "Mrs"
    elif title == "Miss" or title == "Mlle":
        return "Miss"
    else:
        return "Master"
        
```

```python
train_df["New Title"] = train_df["Title"].apply(new_title)
```

```python
train_df = train_df.drop(["Title"], axis = 1)
```

```python
table = train_df.pivot_table(values = "Age", index = ["New Title"], columns = ["Pclass","Sex"],aggfunc = np.mean)
```

```python
def ageFunc(x):
    return table[x["Pclass"]][x["Sex"]][x["New Title"]]
```

```python
train_df["Age"].fillna(train_df[train_df["Age"].isnull()].apply(ageFunc,axis = 1), inplace = True)
```
